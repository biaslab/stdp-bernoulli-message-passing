```python
import matplotlib.pyplot as plt
import numpy as np

# Load saved npz files
and_data = np.load('AND_data.npz')
or_data = np.load('OR_data.npz')
not_data = np.load('NOT_data.npz')

# Replace array conversions with loaded data
snn_output_AND = and_data['snn']
sp_output_AND = and_data['sp']

snn_output_OR = or_data['snn']
sp_output_OR = or_data['sp']

snn_output_NOT = not_data['snn']
sp_output_NOT = not_data['sp']

sp_output_equality = np.load('sum_product_equality.npy')
snn_output_equality = np.load('signal_out_equality.npy')

# Load saved arrays
sp_output_XOR = np.load('sum_product_XOR.npy')
snn_output_XOR = np.load('signal_out_XOR.npy')

```

```python
test_input = [(.3,.4),(.2,.6),(.8,.5),(.7,.3),(.9,.7),(.6,.4),(.5,.1),(.3,.7)]
```

```python
import matplotlib.pyplot as plt

test_input = [(0.3, 0.4), (0.2, 0.6), (0.8, 0.5), (0.7, 0.3),
              (0.9, 0.7), (0.6, 0.4), (0.5, 0.1), (0.3, 0.7)]

# Convert test inputs to string labels
test_labels = [f"{x[0]:.1f},{x[1]:.1f}" for x in test_input]
x_ticks = list(range(len(test_input)))

# Set up the figure and subplots (now 5 plots)
fig, axes = plt.subplots(5, 1, figsize=(10, 16), sharex=True)

# Plot AND
axes[0].plot(snn_output_AND, 'o-', label='SNN', markersize=5)
axes[0].plot(sp_output_AND, 's--', label='Sum-Product', markersize=5)
axes[0].set_title('AND Node Output Comparison')
axes[0].set_ylabel('Output')
axes[0].legend()
axes[0].grid(True)

# Plot OR
axes[1].plot(snn_output_OR, 'o-', label='SNN', markersize=5)
axes[1].plot(sp_output_OR, 's--', label='Sum-Product', markersize=5)
axes[1].set_title('OR Node Output Comparison')
axes[1].set_ylabel('Output')
axes[1].legend()
axes[1].grid(True)

# Plot NOT
axes[2].plot(snn_output_NOT, 'o-', label='SNN', markersize=5)
axes[2].plot(sp_output_NOT, 's--', label='Sum-Product', markersize=5)
axes[2].set_title('NOT Node Output Comparison')
axes[2].set_ylabel('Output')
axes[2].legend()
axes[2].grid(True)

# Plot XOR
axes[3].plot(snn_output_XOR, 'o-', label='SNN', markersize=5)
axes[3].plot(sp_output_XOR, 's--', label='Sum-Product', markersize=5)
axes[3].set_title('XOR Node Output Comparison')
axes[3].set_ylabel('Output')
axes[3].legend()
axes[3].grid(True)

# Plot Equality
axes[4].plot(snn_output_equality, 'o-', label='SNN', markersize=5)
axes[4].plot(sp_output_equality, 's--', label='Sum-Product', markersize=5)
axes[4].set_title('Equality Node Output Comparison')
axes[4].set_xlabel('Test Input (px, py)')
axes[4].set_ylabel('Output')
axes[4].legend()
axes[4].grid(True)
axes[4].set_xticks(x_ticks)
axes[4].set_xticklabels(test_labels, rotation=45)

# Adjust layout
plt.tight_layout()

# Save the figure
plt.savefig("result_all.png")
plt.savefig("result_all.svg")

plt.show()
```

```python
# Extract weight and time data
t_and, w0_and, w1_and = and_data['t'], and_data['w0'], and_data['w1']
t_or,  w0_or,  w1_or  = or_data['t'],  or_data['w0'],  or_data['w1']
t_not, w0_not, w1_not = not_data['t'], not_data['w0'], not_data['w1']

# Create figure with 3 subplots in one row
fig, axes = plt.subplots(1, 3, figsize=(18, 5), sharey=True)

# Plot for AND gate
axes[0].plot(t_and, w0_and, label='$\omega_0$')
axes[0].plot(t_and, w1_and, label='$\omega_1$')
axes[0].set_title('AND Node', fontsize=18)
axes[0].set_xlabel('Time (ms)', fontsize=16)
axes[0].set_ylabel('Synaptic Weight', fontsize=18)
axes[0].grid(True)
axes[0].legend(fontsize=18)
axes[0].tick_params(labelsize=18)

# Plot for OR gate
axes[1].plot(t_or, w0_or, label='$\omega_0$')
axes[1].plot(t_or, w1_or, label='$\omega_1$')
axes[1].set_title('OR Node', fontsize=18)
axes[1].set_xlabel('Time (ms)', fontsize=16)
axes[1].grid(True)
axes[1].legend(fontsize=18)
axes[1].tick_params(labelsize=18)

# Plot for NOT gate
axes[2].plot(t_not, w0_not, label='$\omega_0$')
axes[2].plot(t_not, w1_not, label='$\omega_1$')
axes[2].set_title('NOT Node', fontsize=18)
axes[2].set_xlabel('Time (ms)', fontsize=16)
axes[2].grid(True)
axes[2].legend(fontsize=18)
axes[2].tick_params(labelsize=18)

# Final layout adjustments and save
plt.tight_layout()
plt.savefig("synaptic_weights_all.png")
plt.savefig("synaptic_weights_all.svg")
plt.show()
```
